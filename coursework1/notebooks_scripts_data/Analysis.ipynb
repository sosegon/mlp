# Analysis of results of experiments for coursework 1

In this notebook, the data gathered for the different experiments is analyzed with plots.

The experiments were run in the command line as follows:

- For multiple activations:

        ./mult_act.sh
       
- For multiple hidden layers

        ./mult_layers.sh experiment_name
        
- For multiple initialization strategies

        ./mult_init.sh experiment_name


The first script outputs text files with the following format name: expX_ACTIVATION.txt where X is the number of experiment (related to a a given learning rate) and ACTIVATION is the name of the activation function used to train the model.

The second script outputs text files with the following format name: EXPNAME_selu_layers_X.txt where EXPNAME is the argument experiment_name and X is the number of layers (from 2 to 8) for the model.

The second script outputs text files with the following format name: EXPNAME_selu_INITIALIZATION_layers_2.txt where EXPNAME is the argument experiment_name and INITIALIZATION is the name of the initialization strategy.


All the generated files are located in the same folder of the scripts.

Import libraries

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from collections import OrderedDict
%matplotlib inline
```

## 1. Plots for experiments with different activation functions

### 1.1 Train error, train Accuracy  validation error and validation accuracy

```python
def plot_metrics_for_comparisons(exp_name, lr):
    activations = ["sigmoid", "relu", "lrelu", "elu", "selu"]
    
    # train err | train acc | val err | val acc | params
    metrics_name = ["Train error", "Train Accuracy ", "Validation error", "Validation accuracy", "params"]

    metrics_list = [OrderedDict(), OrderedDict(), OrderedDict(), OrderedDict(), OrderedDict()]

    for activation in activations:
        file_name = "{:s}_{:s}.txt".format(exp_name, activation)
        net_metrics = np.loadtxt(file_name, delimiter=',', skiprows=2)

        assert len(metrics_list) == net_metrics.shape[1]

        for idx, ml in enumerate(metrics_list):
            ml[activation] = net_metrics[:, idx]            

    f, ax = plt.subplots(len(metrics_list) - 1, 1, figsize=(6, 16))
    

    for ml, a, name in zip(metrics_list, ax, metrics_name):
        if name is "params":
            continue
        for net in ml:
            a.plot(np.arange(1, len(ml[net])), ml[net][1:], label=net)
        a.legend(loc=0)
        #a.set_title(name + " | " + hyper)
        a.set_xlabel("Epoch number")
        a.set_ylabel(name)
        a.set_title("Learning rate: {:.3f}".format(lr))
        a.grid()
    
    f.tight_layout()
    f.savefig("{:s}_metrics.svg".format(exp_name), dpi=200)

plot_metrics_for_comparisons("exp0", 0.001)
plot_metrics_for_comparisons("exp1", 0.01)
plot_metrics_for_comparisons("exp2", 0.05)
plot_metrics_for_comparisons("exp3", 0.1)
```

### 1.2 Validation error and training time

```python
def plot_val_err_for_comparisons(exp_names, lrs):
    activations = ["sigmoid", "relu", "lrelu", "elu", "selu"]
    
    # train err | train acc | val err | val acc | params
    metrics_name = ["Validation error"]
    validation_errors = []
    training_times = []

    for exp_name in exp_names:
        currentDict = OrderedDict()
        validation_errors.append(currentDict)
        
        train_time = []
        training_times.append(train_time)
        for activation in activations:
            file_name = "{:s}_{:s}.txt".format(exp_name, activation)
            net_metrics = np.loadtxt(file_name, delimiter=',', skiprows=2)
            currentDict[activation] = net_metrics[:, 2]
            
            with open(file_name, "r") as log_file:
                reader = csv.reader(log_file, delimiter=',')
                train_time.append(float(list(reader)[0][5][7:]))
                

    # Plot validation error
    f, ax = plt.subplots(len(exp_names), 1, figsize=(8, 16))

    for (exp_name, val_errors, a, lr) in zip(exp_names, validation_errors, ax, lrs):
        for net in activations:
            a.plot(np.arange(1, len(val_errors[net])), val_errors[net][1:], label=net)
        a.legend(loc=0)
        a.set_xlabel("Epoch number")
        a.set_ylabel("Validation error")
        a.set_title("Learning rate: {:.3f}".format(lr))
        a.grid()
    
    f.tight_layout()
    f.savefig("activations_val_error.pdf", dpi=200)
    
    # Plot times
    training_times = np.array(training_times).T
    ind = np.arange(len(lrs))
    width = 0.1
    
    f, ax = plt.subplots()
    c = 0
    rects = []
    for tt in training_times:
        rect = ax.bar(np.arange(len(tt)) + c*width, tt, width)
        rects.append(rect)
    
        ax.set_xlabel('Learning rate')
        ax.set_ylabel('Seconds')
        ax.set_title("Training time")
        ax.set_xticks(ind + width*2)
        ax.set_xticklabels(lrs)
        
        c = c + 1
    
    ax.legend(rects, activations)
    f.tight_layout()
    f.savefig("activations_training_time.pdf", dpi=200)

plot_val_err_for_comparisons(["exp0", "exp1", "exp2", "exp3"], [0.001, 0.01, 0.05, 0.1])
```

### 1.3 Evolution of train and evaluation error

```python
def plot_error_difference(exp_name):
    
    activations = ["relu", "lrelu", "elu", "selu"]
    metrics_name = ["Train error", "Validation error", "Error difference"]
    metrics_list = [OrderedDict(), OrderedDict(), OrderedDict()]
        
    for activation in activations:
        file_name = "{:s}_{:s}.txt".format(exp_name, activation)
        net_metrics = np.loadtxt(file_name, delimiter=',', skiprows=2)
        metrics_list[0][activation] = net_metrics[:, 0]
        metrics_list[1][activation] = net_metrics[:, 2]
        metrics_list[2][activation] = net_metrics[:, 2] - net_metrics[:, 0]
    
    f, ax = plt.subplots(1, len(metrics_list), figsize=(24, 8))
    for param, a in zip(metrics_list, ax):
        for net in param:
            a.plot(np.arange(1, len(param[net])), param[net][1:], label=net)
```

```python
plot_error_difference("exp0")
plot_error_difference("exp1")
plot_error_difference("exp2")
plot_error_difference("exp3")
```


### 1.4 Comparison of final training and validation errors

```python
def plot_chart_errors(exp_name, chart_title):
    activations = ["sigmoid","relu", "lrelu", "elu", "selu"]
    metrics_name = ["Train error", "Validation error", "Error difference"]
    train_errors = []
    valid_errors = []
    error_diffs = []
        
    for activation in activations:
        file_name = "{:s}_{:s}.txt".format(exp_name, activation)
        net_metrics = np.loadtxt(file_name, delimiter=',', skiprows=2)
        train_errors.append(net_metrics[-1, 0])
        valid_errors.append(net_metrics[-1, 2])
        error_diffs.append(net_metrics[-1, 2] - net_metrics[-1, 0])
    
    ind = np.arange(len(activations))
    width = 0.20
    
    fig, ax = plt.subplots()
    rects1 = ax.bar(ind, train_errors, width, color='r')
    
    rects2 = ax.bar(ind + width, valid_errors, width, color='b')
    
    rects3 = ax.bar(ind + 2*width, error_diffs, width, color='g')
    
    ax.set_ylabel('Error')
    ax.set_xlabel('Activation')
    ax.set_title(chart_title)
    ax.set_xticks(ind + width)
    ax.set_xticklabels(activations)
    
    ax.legend((rects1[0], rects2[0], rects3[0]), ('Train', 'Validation', "Difference"))
    
plot_chart_errors("exp0", "Learning rate: 0.001")
plot_chart_errors("exp1", "Learning rate: 0.01")
plot_chart_errors("exp2", "Learning rate: 0.05")
plot_chart_errors("exp3", "Learning rate: 0.1")
```

## 2. Plots for experiments with different number of hidden layers

### 2.1 Validation error and accuracy and training time

```python
def plot_val_acc_for_comparisons_mult(exp_name, lr):
    number_layers = range(2, 9)

    validation_accuracies = []
    validation_errors = []
    training_times = []
    
    for net in number_layers:                
        file_name = "{:s}_selu_layers_{:d}.txt".format(exp_name, net)
        net_metrics = np.loadtxt(file_name, delimiter=',', skiprows=2)
        validation_accuracies.append(net_metrics[:, 3])
        validation_errors.append(net_metrics[:, 2])
        
        with open(file_name, "r") as log_file:
            reader = csv.reader(log_file, delimiter=',')
            training_times.append(float(list(reader)[0][5][7:]))

    # Plot accuracy and error for validation
    f, a = plt.subplots(2, 1, figsize=(6, 6))

    for idx, net in enumerate(number_layers):
        a[0].plot(np.arange(1, len(validation_accuracies[idx])), validation_accuracies[idx][1:], label="layers: {:d}".format(net))
        a[0].legend(loc=0)
        a[0].set_title("Learning rate: {:.3f}".format(lr))
        a[0].set_xlabel("Epoch number")
        a[0].set_ylabel("Validation accuracy")
        a[0].grid()
        
        a[1].plot(np.arange(1, len(validation_errors[idx])), validation_errors[idx][1:], label="layers: {:d}".format(net))
        a[1].legend(loc=0)
        a[1].set_title("Learning rate: {:.3f}".format(lr))
        a[1].set_xlabel("Epoch number")
        a[1].set_ylabel("Validation error")
        a[1].grid()

    f.tight_layout()
    f.savefig("validation_acc_err_mult_layers.pdf", dpi=200)

    # Plot times
    ind = np.arange(len(number_layers))
    width = 0.2

    f, ax = plt.subplots()

    rects = []
    rect = ax.bar(ind + width, training_times, width)
    rects.append(rect)

    ax.set_xlabel('Number of hidden layers')
    ax.set_ylabel('Seconds')
    ax.set_title("Training time")
    ax.set_xticks(ind + width)
    ax.set_xticklabels(number_layers)
    
    f.tight_layout()
    f.savefig("training_time_mult_layers.pdf", dpi=200)


plot_val_acc_for_comparisons_mult("exp4_2", 0.001)
```

## 3. Plots for experiments with different initialization approaches

### 3.1 Validation accuracy and error

```python
def plot_val_error_for_comparisons_init(exp_name, lr):
    inits = ["GlorotUniformInit", "SELUInit", "UniformFanInInit", "UniformFanOutInit"]

    validation_errors = []
    validation_accuracies = []
    
    for init in inits:                
        file_name = "{:s}_selu_{:s}_layers_2.txt".format(exp_name, init)
        net_metrics = np.loadtxt(file_name, delimiter=',', skiprows=2)
        validation_accuracies.append(net_metrics[:, 3])
        validation_errors.append(net_metrics[:, 2])

    f, a = plt.subplots(2, 1, figsize=(6, 8))

    for idx, net in enumerate(inits):
        a[0].plot(np.arange(1, len(validation_accuracies[idx])), validation_accuracies[idx][1:], label="{:s}".format(net))
        a[0].legend(loc=0)
        a[0].set_title("Learning rate: {:.3f}".format(lr))
        a[0].set_xlabel("Epoch number")
        a[0].set_ylabel("Validation accuracy")
        a[0].grid()
        
        a[1].plot(np.arange(1, len(validation_errors[idx])), validation_errors[idx][1:], label="{:s}".format(net))
        a[1].legend(loc=0)
        a[1].set_title("Learning rate: {:.3f}".format(lr))
        a[1].set_xlabel("Epoch number")
        a[1].set_ylabel("Validation error")
        a[1].grid()

    f.tight_layout()
    f.savefig("validation_acc_err_init.pdf", dpi=200)

#plot_val_error_for_comparisons_init("exp5_init")
plot_val_error_for_comparisons_init("exp6", 0.001)
```
